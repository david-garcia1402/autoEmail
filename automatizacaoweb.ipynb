```python
import pyautogui as pa
import time
import pandas as pd #biblioteca que permite trabalhar com qualquer tipo de planilha

pa.PAUSE = 1 #vai dar uma pausa de 1 segundo para cada ação

pa.hotkey('ctrl', 't')
pa.write('https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga')
pa.press('enter') #apertar 'enter' para acessar o site
time.sleep(3)
pa.click(x=414, y=333, clicks = 2, button = 'left') #para clicar na pasta exportar
time.sleep(3)
pa.click(x=486, y=383, clicks = 1, button = 'left') #clicar no relatório
time.sleep(2)
pa.click(x=1712, y=222, clicks = 1, button = 'left') #clicar nos 3 pontos
time.sleep(2)
pa.click(x=1522, y=612, clicks = 1, button = 'left') #clicar para fazer download
time.sleep(3) #esperar para baixar o relatório

tabela = pd.read_excel(r'C:\Users\overl\Downloads\Vendas - Dez.xlsx') #comando para o pandas ler uma tabela em excel, mostrando a localização do arquivo em ()
#sempre que passar um caminho de arquivo, colocar o 'r' na frente para impedir os caracteres especiais como o '\'
display(tabela)

faturamento = sum(tabela['Valor Final'])
quantidadeProdutos = sum(tabela['Quantidade'])
display(faturamento)
display(quantidadeProdutos)



pa.hotkey('ctrl', 't') #abrir nova aba
pa.write(r'https://mail.google.com/mail/u/0/#inbox') #digitar o e-mail
pa.press('enter')
time.sleep(2)
pa.click(x=187, y=244, clicks=1, button='left') #clicar para compor o e-mail
time.sleep(2)
pa.write('ov') #escrevendo o e-mail do destinatário
pa.press('enter')
time.sleep(.3)
pa.press('tab') #tab para escrever o assunto do e-mail
time.sleep(2)
pa.write(r'TOTAL DE FATURAMENTO E QUANTIDADE DE PRODUTOS VENDIDOS')
pa.press('tab')
#corpo do e-mail
texto = f"""
Prezados, boa tarde

FATURAMENTO TOTAL: {float(faturamento):,.2f} R$
QUANTIDADE DE PRODUTOS VENDIDOS TOTAL: {quantidadeProdutos:,} Und.

Abrç,
David Garcia Comercial
"""
pa.write(texto)
pa.hotkey('ctrl', 'enter') #enviar o e-mail
```
